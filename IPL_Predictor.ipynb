## Getting wickets

```python
import numpy as np
import pandas as pd

# The original 'player_dismissed' column contains NaN values for non-dismissals.
# Convert NaN to 0 (not dismissed) and any non-NaN value to 1 (dismissed).
second_innings['player_dismissed'] = np.where(pd.isna(second_innings['player_dismissed']), 0, 1).astype(int)

# Calculate cumulative wickets taken for each match.
# Using .transform('cumsum') correctly aligns the cumulative sum back to the original DataFrame.
wickets = second_innings.groupby('match_id')['player_dismissed'].transform('cumsum')

# Calculate wickets remaining (10 total wickets minus current cumulative wickets).
second_innings['wickets_remaining'] = 10 - wickets

# Displaying the head to show the new 'wickets_remaining' column
print("Updated second_innings DataFrame with 'wickets_remaining':")
print(second_innings.head())
```

### Current Run Rate

```python
second_innings['crr'] = round((second_innings['current_Score']*6)/(120 - second_innings['balls_left']),2)
second_innings
```

### Required Run Rate

```python
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
```

### Sampling Data

```python
# Re-defining the winner function and ensuring 'results', 'required_data', and 'final_data' are properly created.

def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0

# Ensure the 'results' column is created in second_innings
# This step re-applies the winner function to create the 'results' column,
# resolving the KeyError observed previously.
second_innings['results'] = second_innings.apply(winner, axis=1)

# Re-creating 'required_data' as it was not defined due to previous errors.
# This selects the necessary columns from the updated second_innings DataFrame.
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','total_runs_x','crr','rrr','results']]

# The original operation for creating 'final_data' is now executed here.
# This ensures 'final_data' is defined and available for subsequent cells.
final_data = required_data.sample(required_data.shape[0])
```

### Dropping NUll Values

```python
import numpy as np
final_data.replace([np.inf, -np.inf], np.nan, inplace=True)
final_data.dropna(inplace=True)
```

### Removing rows where balls_left==0 to remove -inf and inf from rrr column

```python
final_data = final_data[final_data['balls_left']!=0]
```

### Storing data in new csv

```python
final_data.to_csv("final_data.csv", index=False)
```

### Building Model

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:, :-1]
y = final_data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.25,random_state=100)
x_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```

## Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
trf = ColumnTransformer(
    [('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])],
    remainder='passthrough'
)

pipe = Pipeline([
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

accuracy_score(y_test,y_pred)
```
